```python
# 문자열은 어떻게 만들고 사용할까?

## 1. 큰 따옴표로 양쪽 둘러싸기
a = "Hello World"

## 2. 작은 따옴표로 양쪽 둘러싸기
b = 'Phtyon is fun'

## 3. 큰 따옴표 3개로 둘러싸기
c = """Life is too short, You need Python"""

## 4. 작은 따옴표 3개로 둘러싸기
d = '''Life is too short, You need Python'''

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열 안에 작은 따옴표를 포함하기
a = "python's favorit food is perl"
b = """python's favorit food is perl"""
c = 'python\'s favorit food is perl'
print(a)
print(b)
print(c)
```

```python
# 문자열 안에 큰 따옴표 표함하기
a = 'Python is very easy" he says.'
b = "\"Python is very easy\" he says."

print(a)
print(b)
```

```python
multiline = """Life is too short, 
you need python."""
multiline2 = "Life is too short, \nyou need python."
print(multiline)
print(multiline2)
```

```python
# 문자열 포매팅
## 1. 숫자 바로 대입

print("I eat %d apples" % 3)

## 2. 문자열 바로 대입
print("I eat %s apple" % "five")
```

```python
## 3. 숫자값을 나타내는 변수로 대입
Number = 3
print("I eat %d apples" % Number)
```

```python
## 4. 두 개 이상의 값 넣기

Number = 10
day = "three"

print("I eat print %d apples so I was sick for %s days" % (Number, day))
```

```python
# foramt 함수를 사용한 포매팅

print(" I eat {0} apples".format(3))
print(" I eat {0} apples".format("five"))
```

```python
# format 함수를 이용해서 두 개 이상의 값 넣기

Number = 7
day = "three"
print("I eat print {0} apples so I was sick for {1} days".format(Number, day))
```

```python
# f문자열 포매팅

name = "홍길동"
age = 10
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
# 문자열에 관련된 함수
a = "hobby"
print(type(a))

## 문자의 개수 세기 - count
print(a.count("b"))
```

```python
# 위치 알려주기 - find
a = "Python is the best choice"
print(a.find("b"))
print(a.find("k"))  #없으면 -1 값 반영
```

```python
# 위치 알려주기 두 번째 - index

a = "Life is too short"
print(a.index("t")) #젤 처음에 나온 값만 출력
print(a.index("k"))
```

```python
# 문자열 삽입 - join

print(",".join("abcd"))
```

```python
# 소문자를 대문자로 바꾸기 - upper

a = "hi"
print(a.upper())
```

```python
# 대문자를 소문자로 바꾸기 - lower

a = "HI"
print(a.lower())
```

```python
# 공백 지우기 - strip
a = "   hi   "
print(a)  
a = a.strip()  # a 자체는 변하지 않음
print(a)
```

```python
# 공백 지우기 왼쪽 / 오른쪽
a = "   hi   "
print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기- replace
a = "Life is too short"
a.replace("Life","your leg")
```

```python
# 문자열 나누기 - split
a = "Life is too short"
a.split()  # 괄호안의 값을 기준으로 분리하여서 리스트 형태로 출력
```

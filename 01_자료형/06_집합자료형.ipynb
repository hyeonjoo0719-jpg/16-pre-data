```python
# 집합 자료형은 어떻게 만들까?
s1 = set([1, 2, 3])
s2 = {4, 5, 6}

print(s1)
print(type(s1))
print(s2)
print(type(s2))
```

```python
s3 = set("hello")  # 각자의 캐릭터가 하나의 요소로 들어가고 중복을 허락하지 않음, 순서가 없음
print(s3)
```

```python
# 집합자료형의 특징
## 1. 중복을 허용하지 않습니다.
## 2. 순서가 없다.
```

```python
# 교집합 구하기
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print(s1)
print(s2)

## 1. intersection()
print(s1.intersection(s2))  # s2.intersection(s1) 과 같음

## 2. &
print(s1 & s2)
```

```python
# 합집합 구하기
## 1. union()
print(s1.union(s2))

## 2. | (파이프)
print(s1 | s2)
```

```python
# 차집합 구하기
## 1. difference()
print(s1.difference(s2))  # s1 - s2
print(s2.difference(s1))  # s2 - s1

## 2. 빼기 연산자
print(s1 - s2)
print(s2 - s1)
```

```python
# 집합 자료형 관련 함수
## 1. 값을 한 개 추가하기 - add
s1 = {1, 2, 3}
s1.add(4)
print(s1)

## 2. 값 여러 개 추가하기 - update
s1 = {1, 2, 3}
s1.update([4, 5, 6])
print(s1)

## 3. 특정 값 제거하기 - remove
s1 = {1, 2, 3}
s1.remove(2)
print(s1)

## 4. 모든 값 제거하기 - clear
s1 = {1, 2, 3}
s1.clear()
print(s1)
```

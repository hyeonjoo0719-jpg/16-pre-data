```python
print(1 == 1)
print(2 > 1)  # 왼쪽이 주어(기준)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2)
```

```python
# if 0:
#     수행할 문장
print(bool("python"))
print(bool(""))  # 문자열에서는 없는 값을 제외하고는 모두 Ture로 판단
print(bool(" "))

print(bool(10))
print(bool(0))   # 숫자열에서는 0빼고는 모두 True로 판단

print(bool([1, 2, 3]))
print(bool([]))  # 리스트는 하나라도 있으면 True 아무것도 없으면 False

print(bool({"name" : "pey"}))
print(bool({}))  # 딕셔너리도 아무것도 없으면 False
```

```python
# 논리 연산자
## 1. and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)

## 2. or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)  #alt 누른 상태로 클릭하면 다중 선택 가능
```

```python
## 3. not 연산자
print(not True)
```

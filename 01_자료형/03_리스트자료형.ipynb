```python
# 리스트는 어떻게 만들고 사용할까?

# 리스트를 만드는 형식 --> 리스트명 = [요소1, 요소2, ...]

a = []   #비어있는 리스트
b = [1, 2, 3]
c = ["Life", "is", "too", "short"]
d = [1, 2, "Life", "is"]
e = [1, 2, ["Life", "is"]]  #리스트 안에 리스트도 가능 요거는 세 가지 요소

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱
a = [1, 2, 3]
print(a[0])
print(a[2])
print(type(a[2])) # a 는 리스트이지만 각각의 요소는 각자의 속성을 가지고 있음 
print(a[0]+a[2])
print(a[-1])

b = ['1', '2', '3', 'a', 'b', 'c']
print(b[-1])
print(b[0] + b[2] + b[-1])

c = [1, 2, ['a', 'b', ["Life", "is"]]]
print(len(c))
print(c[-1])
print(c[-1][-1])
print(c[-1][-1][0])
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
print(a[0:2])
print(a[0:0])
print(a[2:])
print(a[:3])

a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]
print(a[3][1:])
print(type(a[3][1:]))  #슬라이싱을 가져오면 각각의 속성이 아니고 리스트 형태를 가짐
```

```python
# 리스트 더하기(+)
a = [1, 2, 3]
b = [4, 5, 6]

print(a + b)
```

```python
# 리스트 곱하기(*)
a = [1, 2, 3]
print(a * 3)
```

```python
# 리스트의 수정
a = [1, 2, 3]
a[2] = 4  
print(a)
```

```python
# 리스트의 요소를 삭제하기
a = [1, 2, 3]
del a[1]
print(a)

a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트에 관련된 함수
## 리스트 요소 추가하기 - append
a = [1, 2, 3]
a.append(4)
print(a)
a.append([5, 6])
print(a)
```

```python
# 리스트 정렬 - sort
a = [1, 4, 2, 3]
a.sort()    # 오름차순 정렬  , 요 sort와 같은 걸 매서드 라고 함
print(a)    
a.sort(reverse=True)    # 내림차순 정렬
print(a)
```

```python
a = [1, 4, 2, 3]
print(sorted(a))   # 요거는 내장함수
print(sorted(a, reverse=True)) #기본이 False이고, True로 하면 내림차순
```

```python
# 리스트 뒤집기 - reverse
a = ['a', 'c', 'b']
a.reverse()
print(a)  
```

```python
# 인덱스 반환 - index
a = ['a', 'c', 'b']
print(a.index("c"))
```

```python
# 리스트에 요소 삽입 - insert
a = [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
# 리스트에 요소 제거 - remove
a = [1, 2, 3, 1, 2, 3]
a.remove(3)  # 가장 처음 찾아진 객체만 지워줌
print(a)
a.remove(3)
print(a)
```

```python
# 리스트의 요소 꺼내기
a = [1, 2, 3]
result = a.pop()  # 기본이 인덱스가 -1이라서 마지막 요소를 뽑아내서 지움
print(result)     # 뽑아낸 객체
print(a)
```

```python
# 리스트에 포한된 요소 x의 개수 세기 - count
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장 - extend
a = [1, 2, 3]
a.extend([4, 5])  #아까 append는 리스트가 하나의 리스트 형태로 들어갔는데 이거는 결합되는 형태 + 연산과 같음
print(a)
```

```python
# 튜플은 어떻게 만들까? (tuple)
t1 = ()   # 비어있는 튜플
t2 = (1, )  # 튜플은 수정하거나 삭제할 수 없음 // 하나의 요소를 적을 때도 반드시 콤마를 써줘야 함
t3 = (1, 2, 3)  # 2개 이상 요소일 때 마지막에는 콤마가 없어도 됨
t4 = 1, 2, 3    # 튜플은 괄호 생략 가능
t5 = ('a', 'b', ('ab', 'cd')) #튜플 안에 문자열도 가능하고, 튜플 안에 튜플도 가능

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))

print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```

```python
a = 1,      # 튜플의 하나의 요소에서 콤마를 안찍어주면 int로 인식됨
print(type(a))
```

```python
# 튜플의 요소 값을 지우려고 하면 어떻게 될까?

t1 = (1, 2, 'a', 'b')
del t1[0]    # TypeError: 'tuple' object doesn't support item deletion --> 삭제 안됨
```

```python
# 튜플의 요소 값을 변경하려고 하면 어떻게 될까?
t1 = (1, 2, 'a', 'b')
t1[0] = 'c'    # TypeError: 'tuple' object does not support item assignment --> 변경 안됨
```

```python
# 튜플 인덱싱
t1 = (1, 2, 'a', 'b')
print(t1[2])
print(t1[-1])
```

```python
# 튜플 슬라이싱
t1 = (1, 2, 'a', 'b')
t1[1:]
```

```python
# 튜플 더하기
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
print(t1 + t2)  # 결합이 됨
print(t1)  # 기존 t1은 변하지 않음
```

```python
# 튜플 곱하기
t2 = (3, 4)
print(t2*3) # 3번 반복
print(t2)   # 기존 t2는 변하지 않음
```

```python
# 튜플 길이 구하기
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
print(len(t1))
print(len(t2))
```

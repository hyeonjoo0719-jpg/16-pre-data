```python
# 딕셔너리는 어떻게 만들까?
# {key1: value1,key2: value2, ...}
dic ={
    'name' : 'pey',
    'phone' : '010-1234-1234',
    'birth' : '1118'
}
print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기
a = { 1 : 'a'}
a[2] = 'b'  # 인덱스 번호가 아니고 key임
print(a)
a[3] = [1, 2, 3]
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)
```

```python
# 딕셔너리에 key를 사용해서 value 얻기
grade = {
    'pey' : 10,
    'juli' : 99
}

print(grade['pey'])
print(grade['juli'])
```

```python
# 딕셔너리 관련 함수
## 1. key 리스트 만들기 - keys

a = {
    'name' : 'pey',
    'phone' : '010-1234-1234',
    'birth' : '1118'
}
list(a.keys()) #key에 해당하는 값들 출력
```

```python
## 2. value 리스트 만들기 - values
a.values()  #value에 해당하는 값들 출력
```

```python
## 3. key와 value 쌍으로 얻기 - items
print(a)
a.items()
list(a.items())
```

```python
## 4. key와 value의 쌍 모두 지우기
a.clear()
print(a)
```

```python
# key로 value 얻기 - get
a = {
    'name' : 'pey',
    'phone' : '010-1234-1234',
    'birth' : '1118'
}
print(a['name'])
print(a.get('name'))
```

```python
# a['hobby']
a.get('hobby',0) # 아무것도 없을 때는 0을 반환
```

```python
# 해당 key가 딕셔너리 안에 있는 지 확인 - in
print(a)
print('name' in a)
print('email' in a)
```

```python
# key로 value 꺼내기 -pop
a = {
    'name' : 'pey',
    'phone' : '010-1234-1234',
    'birth' : '1118'
}
result = a.pop('phone')
print(result)
print(a)
```

```python
# 딕셔너리의 키값으로 올 수 없는 데이터 타입은 리스트 --> mutable
a = [1, 2, 3]
a[0] = 3
print(a)

```

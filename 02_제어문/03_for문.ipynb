```python
# for문의 기본 구조
# for 변수 in 리스트(또는 튜플, 문자열):  #in 다음에는 꺼내쓸 수 있는 것이 옴 ,, 이 변수가 있는 동안에 반복
#     수행할 문장1
#     수행할 문장2
```

```python
test_list = ['one', 'two', 'tree']

for tl in test_list:
    print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)]  #튜플형태를 가진 3개의 요소


# for item in a:
#     print(item)

for (k, j) in a:  # 튜플형태로도 가능
    print(k, j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0  # 임의의 수를 할당
for mark in marks:
    number += 1
    if mark >= 60:
        print(f"{number}번 학생은 합격입니다.")
    else:
        print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문과 continue

marks = [90, 25, 67, 45, 80]
number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue
    
    print(f"{number}번 학생은 합격힙니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)
list(a)

b = range(1, 11, 2)   # 슬라이싱과 같이 지정한 자리 -1까지 출력, 세번째 인수를 넣어주면 step 적용
list(b)

c = range(10, 0, -1)
list(c)
```

```python
# for와 range를 이용한 구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i} x {j} = {i * j}")
```

```python
# 리스트 컴프리헨션 사용하기
a = [1, 2, 3, 4]
result = []
for num in a:
    result.append(num*3)
print(result)

a = [1, 2, 3, 4]
result = [num * 3 for num in a]   # 리스트 컴프리헨션
print(result)
```

```python
# for문과 break문
for i in range(10):
    if i == 5:
        break        # 컨티뉴와는 다르게 5와 같아지면 중지됨
    else:
        print(i)
```

```python
# enumerate() 함수
fruits = ["apple", "banana", "orange"]
for fruit in enumerate(fruits):            # enumerate는 튜플 형태로 가져옴
    print(f"과일: {fruit}")


fruits = ["apple", "banana", "orange"]
for i, fruit in enumerate(fruits):            
    print(f"{i}번째 과일: {fruit}")
```

```python
# zip 함수로 여러 리스트 순회하기
names = ["홍길동", "김철수", "이영희"]
scores = [85, 92, 78]

for item in zip(names, scores):    # 붙여서 튜플 형태로 출력
    print(item)

for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```

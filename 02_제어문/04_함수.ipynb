```python
#1. 일반적인 함수
## 매개변수가 o, 반환값 o

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...
#     return 반환값

def add(a, b):
    result = a+b
    return result

a = add(3, 4)
print(a)
```

```python
#2. 입력값이 없는 함수
## 매개변수 X, 반환값 o

def say():
    return "Hi"

a = say()
print(a)
```

```python
#3. 반환값이 없는 함수
## 매개변수 o, 반환값 X(retur=None)

def add(a,b):
    print(f"{a}, {b}의 합은 {a}+{b}")
   #return None 이 안적으면 숨겨져 있음
a = add(3,4) 
print(a)          # 리턴값이 없기 때문에 a에 아무것도 할당되지 않음
```

```python
#4. 입력값도 없고 반환값도 없는 함수
##4. 매개변수 X, 반환값 X
def say():
    print("hi")
   #return None 이 숨겨져 있음
a = say()
print(a)
```

```python
# 매개변수를 지정하여 호출하기

def sub(a, b):
    result = a - b
    return result

c = sub(a=7, b=3)
print(c)

d = sub(b=7, a=3)
print(d)
```

```python
# 입력 값이 몇 개가 될 지 모를때는 어떻게 해야 할까?

# def 함수이름(*매개변수):
#     수행할 문장...

def add_many(*args):     # 앞에 *을 적으면 인수가 몇 개가 될 지 모른다는 의미(근데 파이썬에서 암묵적으로 *args라고 씀)
    result = 0
    print(args)

    for i in args:
        result += i
    return result
c = add_many(1, 2, 3, 4, 5)
print(c)

```
